Creating student portal that includes multiple features as adding new student with age subject class , having access to mange file append studnet profile subjects and his/her Information.

```python

import csv  # Used for reading/writing CSV files
students = []   #  Initializing student[] as an master list to store student record.
def load_from_file(filename="students.csv"):  #declearing function as an load from file
    """
    Loads student records from a CSV file.
    Converts string values to appropriate types using eval.
    Handles file not found and other errors.
    """
    try:
        with open(filename, mode='r') as file:
            reader = csv.DictReader(file)
            for row in reader:
                row['age'] = int(row['age'])  # Convert age from string to int
                row['subject_list'] = eval(row['subject_list'])  # Convert string to list
                row['marks'] = eval(row['marks'])  # Convert string to dictionary
                students.append(row)
        print(" Records loaded successfully.")
    except FileNotFoundError:
        print(" No previous data found. Start Again.")
    except Exception as e:
        print(" Error loading file:", e)

def save_to_file(filename="students.csv"):  #Declaring function as an Save to file
    """
    Saves current student records to a CSV file.
    Uses DictWriter to store structured data.
    """
    try:
        with open(filename, mode='w', newline='') as file:
            fieldnames = ['name', 'age', 'class', 'subject_list', 'marks']
            writer = csv.DictWriter(file, fieldnames=fieldnames)
            writer.writeheader()
            for student in students:
                writer.writerow(student)
        print(" Data saved to file.")
    except Exception as e:
        print(" Error saving file:", e)
def add_student():   #Declaring Function as an Add student
    """
    Takes user input to add a new student.
    Stores as a dictionary and appends to master list.
    """
    try:
        name = input("Enter name: ")
        age = int(input("Enter age: "))
        class_name = input("Enter class: ")
        subjects = input("Enter subjects (use , to seprate subject): ").split(',')
        subjects = [sub.strip() for sub in subjects]  # Clean whitespace
        marks = {}
        for subject in subjects:
            marks[subject] = int(input(f"Enter marks for {subject}: "))
        student = {
            'name': name,
            'age': age,
            'class': class_name,
            'subject_list': subjects,
            'marks': marks
        }
        students.append(student)
        print(" Student added.")
    except Exception as e:
        print(" Error adding student:", e)

def view_all():   #Declaring function as an View all Records
    """
    Displays all student records nicely.
    """
    if not students:
        print(" No records found.")
        return
    for student in students:
        print("\n                     ")
        print(f"Name   : {student['name']}")
        print(f"Age    : {student['age']}")
        print(f"Class  : {student['class']}")
        print(f"Subjects: {', '.join(student['subject_list'])}")
        print(f"Marks  : {student['marks']}")

def search_by_name(): #Declaring function as an Search by name
    """
    Searches for a student by name using a loop.
    """
    name = input("Enter name to search: ")
    found = False
    for student in students:
        if student['name'].lower() == name.lower():
            print(" Student found:")
            print(student)
            found = True
            break
    if not found:
        print(" Student not found.")

def update_student(): #Declaring function as an Update Information
    """
    Finds a student by name and updates age and marks.
    """
    name = input("Enter name to update: ")
    for student in students:
        if student['name'].lower() == name.lower():
            try:
                student['age'] = int(input("Enter new age: "))
                for subject in student['subject_list']:
                    student['marks'][subject] = int(input(f"New marks for {subject}: "))
                print(" Student updated.")
            except Exception as e:
                print(" Invalid Entry:", e)
            return
    print(" Student not found.")

def delete_student():# Declaring Functon as Delete a Record
    """
    Deletes a student by name using loop and remove.
    """
    name = input("Enter name to delete: ")
    for i, student in enumerate(students):
        if student['name'].lower() == name.lower():
            del students[i]  # Remove by index
            print(" Student deleted.")
            return
    print(" Student not found.")

def analyze_grades(): #Declaring Function as an Analyzing grades
    """
    Calculates and displays how many students got A, B, etc.
    Also shows the average of all students.
    """
    if not students:
        print(" No data to analyze.")
        return

    grade_count = {'A': 0, 'B': 0, 'C': 0, 'D': 0, 'F': 0}
    total_avg = 0

    for student in students:
        marks = student['marks'].values()
        avg = sum(marks) / len(marks)
        total_avg += avg

        # Assign grade based on average(Grading system)
        if avg >= 90:
            grade_count['A'] += 1
        elif avg >= 80:
            grade_count['B'] += 1
        elif avg >= 70:
            grade_count['C'] += 1
        elif avg >= 60:
            grade_count['D'] += 1
        else:
            grade_count['F'] += 1

    print("\n Grade Analysis:")
    for grade, count in grade_count.items():
        print(f"{grade}: {count} student(s)")
    print("Average marks across all students:", round(total_avg / len(students), 2))

def list_unique_subjects(): #Declaring Function as an list of Unique subjects
    """
    Uses set to extract unique subjects across all students.
    """
    all_subjects = set()
    for student in students:
        all_subjects.update(student['subject_list'])
    print(" Unique Subjects:", ", ".join(all_subjects))

def main():  #Main Search menu for different features
    """
    The main menu that runs the whole program in a loop.
    Handles invalid inputs with try/except.
    """
    load_from_file()  # Load data at the start
    while True:
        print("\n Welcome to  School Portal Menu")
        print("1. Add Student Record")
        print("2. View All Records")
        print("3. Search by Name")
        print("4. Update Student Info")
        print("5. Delete a Record")
        print("6. Analyze Grades")
        print("7. List Unique Subjects")
        print("8. Save to File")
        print("9. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            add_student()
        elif choice == '2':
            view_all()
        elif choice == '3':
            search_by_name()
        elif choice == '4':
            update_student()
        elif choice == '5':
            delete_student()
        elif choice == '6':
            analyze_grades()
        elif choice == '7':
            list_unique_subjects()
        elif choice == '8':
            save_to_file()
        elif choice == '9':
            save_to_file()  # Save before exiting
            print(" Goodbye!")
            break
        else:
            print(" Invalid choice, try again.")
main()
#Go back to Portal
```
